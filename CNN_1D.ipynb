```python
import torch
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

# Importer les fonctions du module
import sys
sys.path.append('/content/drive/MyDrive/VO_TARVERDIAN')
from utils import cnn_1d_model as cd
# Vérification de GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using: {device}")

# Constantes et chemins
AUDIO_DIR ="/content/drive/MyDrive/VO_TARVERDIAN/data/data/X_train"
LABELS_FILE = "/content/drive/MyDrive/VO_TARVERDIAN/data/data/Y_train_ofTdMHi.csv"
TEST_DIR = "/content/drive/MyDrive/VO_TARVERDIAN/data/data/X_test"
MODEL_SAVE_PATH = "/content/drive/MyDrive/VO_TARVERDIAN/final_CNN/best_cetacean_model_modified.pt"
PREDICTION_FILE ="/content/drive/MyDrive/VO_TARVERDIAN/final_CNN/predictions.csv"
# Paramètres
SAMPLE_RATE = 256000
AUDIO_DURATION = 0.20
TARGET_LENGTH = int(SAMPLE_RATE * AUDIO_DURATION)
BATCH_SIZE = 16

# 1. Explorer les fichiers audio
cd.explore_audio_files(AUDIO_DIR)

# 2. Charger et préparer les données
labels_df = pd.read_csv(LABELS_FILE)
labels_df['file_path'] = labels_df['id'].apply(lambda x: os.path.join(AUDIO_DIR, x))

# 3. Créer les DataLoaders
train_loader, valid_loader = cd.prepare_data_loaders(
    labels_df, TARGET_LENGTH, SAMPLE_RATE, BATCH_SIZE
)

# 4. Initialiser le modèle
test_audio = np.zeros(TARGET_LENGTH)
test_features = cd.extract_click_features(test_audio, SAMPLE_RATE)
num_features = len(test_features)

model = cd.CNN1DModel(TARGET_LENGTH, click_features_size=num_features).to(device)

# 5. Entraîner le modèle
trained_model, history = cd.train_model(
    model,
    train_loader,
    valid_loader,
    num_epochs=25,
    lr=0.001,
    patience=5,
    save_path=MODEL_SAVE_PATH
)

# 6. Tracer les courbes d'apprentissage
cd.plot_learning_curves(
    history['train_loss'],
    history['valid_loss'],
    history['train_acc'],
    history['valid_acc']
)

# 7. Évaluer le modèle
cd.evaluate_model_detailed(trained_model, valid_loader)

# 8. Préparer les données de test et faire des prédictions
test_loader = cd.prepare_test_loader(TEST_DIR, TARGET_LENGTH, SAMPLE_RATE, BATCH_SIZE)
predictions_df = cd.predict_test(trained_model, test_loader)
predictions_df.to_csv(PREDICTION_FILE, index=False)
print(f"Prédictions sauvegardées dans {PREDICTION_FILE}")
```
